```python
df["km_driven"]=df["km_driven"].fillna(df["km_driven"].mean())
```

```python
df=df.dropna()
```

```python
df=df.drop("name",axis=1)
df
```

```python
df.corr(numeric_only=True)
```

```python
plt.figure()
plt.hist(df["selling_price"])
plt.xlabel("Selling price of cars")
plt.ylabel("frequency")
plt.title("frequency of selling price")
plt.show()


```

 Inferance:
 1)Most cars fall in lower price range
 2)Right-skewed distribution
 3)Few high-priced cars (luxury segment)

```python
plt.figure()
plt.scatter(df["year"],df["selling_price"])
plt.xlabel("YEAR")
plt.ylabel("selling_price")
plt.title("YEAR VS SELLING PRICE")
plt.show()
```

Inferance:
1)Positive relationship between year and price
2) Newer cars are more expensive
3) Some outliers exist

```python
plt.figure()
plt.bar(df["fuel"],df["selling_price"])
plt.xlabel('Fuel Types')
plt.ylabel("Selling Price")
plt.title("Fuel Type vs Selling Price")
plt.show()
```

Inferance:
1) Petrol and Diesel dominate dataset
2) Price varies across fuel types
3) Not aggregated → appears cluttered

Fuel type affects price but visualization is raw

```python
plt.figure()
plt.pie(df["owner"].value_counts(),labels=df["owner"].unique(),startangle=90,autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

Inferance:
1) Majority cars are First Owner
2) Very few high-owner vehicles
3) Indicates resale pattern

```python
plt.figure()
plt.boxplot(df["year"])
plt.ylabel("YEAR")
plt.title("BOXPLOT")
plt.show()
```

Inferance:
1)Shows spread of manufacturing years
2) Most cars are recent
3) Some outliers (very old cars)

```python
plt.figure()
plt.stem(df["year"],df["selling_price"])
plt.xlabel("YEAR")
plt.ylabel("selling_price")
plt.title("STEMPLOT")
plt.show()
```

Inferance:
1)Shows discrete relationship
2) Confirms trend: newer → higher price
3) Useful for visualizing spikes

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

Inferance:
1)Shows relationships between numerical variables

2)Year vs Selling Price → positive
3)Km driven vs Price → negative (generally)

4) Helps in feature importance understanding

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.violinplot(x="fuel", y="selling_price", data=df)
plt.xlabel("Fuel Type")
plt.ylabel("Selling Price")
plt.title("Fuel Type vs Selling Price Distribution (Violin Plot)")
plt.show()
```

Inferance:
1) Fuel type has a significant impact on selling price
2) Diesel cars tend to have higher price distribution
3) Petrol cars dominate but mostly in lower price range
4) CNG and LPG cars are generally cheaper options
5) Electric vehicles are underrepresented in the dataset
