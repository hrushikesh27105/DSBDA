```python
df['km_driven']=df['km_driven'].fillna(df['km_driven'].mean())
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df=df.drop('name',axis=1)
df
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.corr(numeric_only=True)
```

```python
df.corr()
```

```python
x=df[["transmission_Manual"]]
y=df['selling_price']
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse=root_mean_squared_error(y_test,y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
transmission = input("Enter transmission (Manual/Automatic): ")

if transmission.lower() == "manual":
    value = 1
elif transmission.lower() =="Automatic":
    value = 0
else:
    print("Invalid input. Please enter 'Manual' or 'Automatic'.")



prediction = reg.predict([[value]])

print("Predicted Selling Price:", prediction[0])
```

```python
x1=df.drop('selling_price',axis=1)
y1=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,train_size=0.7)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x1_train,y1_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y1_pred=lr.predict(x1_test)
y1_pred
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(y1_test,y1_pred)
mae=mean_absolute_error(y1_test,y1_pred)
r2=r2_score(y1_test,y1_pred)
rmse=root_mean_squared_error(y1_test,y1_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
print(f"Root Mean Squared Error: {rmse}")
```
